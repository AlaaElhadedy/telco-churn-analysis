```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv(r"C:\Users\ME\Downloads\archive (11)\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# تحويل TotalCharges لـ numeric (هيحول المسافات لـ NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# دلوقتي هتظهر القيم المفقودة
print(df.isnull().sum())

# علاج بالوسيط
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())

print("✅ تم علاج القيم المفقودة")
```

```python
# رسم BoxPlot لاكتشاف القيم الشاذة
plt.figure(figsize=(12, 4))

cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(f'Outliers في {col}')

plt.tight_layout()
plt.show()
```

```python
# معالجة بطريقة IQR
for col in ['tenure', 'MonthlyCharges', 'TotalCharges']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)

print("✅ تم معالجة القيم الشاذة")
print(df[['tenure', 'MonthlyCharges', 'TotalCharges']].describe())
```

```python
# شوف التوزيع
print(df['Churn'].value_counts())
print(df['Churn'].value_counts(normalize=True) * 100)

# رسم التوزيع
sns.countplot(x='Churn', data=df, palette='Set2')
plt.title('توزيع Churn')
plt.show()
```

```python
# تحويل الأعمدة النصية لـ numeric
le = LabelEncoder()
df_encoded = df.copy()

for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# تقسيم X و y
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

print("قبل SMOTE:", y.value_counts().to_dict())

# تطبيق SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("بعد SMOTE:", pd.Series(y_resampled).value_counts().to_dict())
print("✅ البيانات بقت متوازنة")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, 
    test_size=0.2, 
    random_state=42
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
print("✅ تم تقسيم الداتا")
```

```python
# تدريب النموذج
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)

# تقييم النموذج
y_pred_lr = lr_model.predict(X_test)

print("=== Logistic Regression ===")
print(classification_report(y_test, y_pred_lr))

# Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred_lr), 
            annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

```python
# تدريب النموذج
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)

# تقييم النموذج
y_pred_rf = rf_model.predict(X_test)

print("=== Random Forest ===")
print(classification_report(y_test, y_pred_rf))

# Confusion Matrix
sns.heatmap(confusion_matrix(y_test, y_pred_rf), 
            annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix - Random Forest')
plt.show()
```

```python
# مقارنة F1 Score
lr_f1 = f1_score(y_test, y_pred_lr)
rf_f1 = f1_score(y_test, y_pred_rf)

print("=== مقارنة النموذجين ===")
print(f"Logistic Regression F1: {lr_f1:.2f}")
print(f"Random Forest F1:       {rf_f1:.2f}")

# رسم المقارنة
models = ['Logistic Regression', 'Random Forest']
scores = [lr_f1, rf_f1]

sns.barplot(x=models, y=scores, palette='Set2')
plt.title('مقارنة F1 Score')
plt.ylim(0, 1)
plt.show()

print(f"\n✅ أفضل نموذج: {'Random Forest' if rf_f1 > lr_f1 else 'Logistic Regression'}")
```

```python
# أهم العوامل في النموذج
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(feature_importance.head(10))

# رسم أهم 10 عوامل
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', 
            data=feature_importance.head(10), 
            palette='viridis')
plt.title('أهم 10 عوامل في توقع Churn')
plt.tight_layout()
plt.show()
```

```python
# تصدير النتائج
# 1. البيانات الأساسية
df.to_csv('churn_clean.csv', index=False)

# 2. Feature Importance
feature_importance.to_csv('feature_importance.csv', index=False)

# 3. نتائج النموذجين
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'F1_Score': [round(lr_f1, 2), round(rf_f1, 2)]
})
results.to_csv('model_results.csv', index=False)

print("✅ اتصدرت 3 ملفات:")
print("   - churn_clean.csv")
print("   - feature_importance.csv")
print("   - model_results.csv")
```
